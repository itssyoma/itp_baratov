Класс «Куб». Реализовать ввод и вывод полей данных, вычисление объема, площади поверхности, длины диагонали, а также вывод информации об объекте

2. Объявление класса

```python
class Cube:

  def __init__(self, length):
    self.a = float(length)
  
  def GetV(self):
    return self.a**3
  
  def GetS(self):
    return 6*self.a*self.a

  def GetD(self):
    return round(self.a*(3**0.5), 2)
  
  def Info(self):
    print('\n'+'*'*20)
    print('Куб: \n')
    print(f'Длина стороны: {self.a}')
    print(f'Длина диагонали: {self.GetD()}')
    print(f'Площадь поверхности: {self.GetS()}')
    print(f'Объем: {self.GetV()}')
    print('*'*20)
```

3. Демонстрация использования класса
(создание объекта и вызов методов)

```python
a1 = Cube(input('Введите длину стороны куба\n'))
a1.Info()
```
