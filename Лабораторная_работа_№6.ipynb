Создадим фильтр

```python
from PIL import Image
from PIL import ImageFilter
with Image.open('cat.jpeg') as img:
  img.load()
print("Изображение без фильтра:")
img.show()

def myFilter(img):
  converted_img = img.transpose(Image.FLIP_LEFT_RIGHT).crop((60, 50, 500, 380))
  kernel3x3 = [-8,-11,9,
               13,7,8,
               -1,-2,-2]
  
  scale = None
  offset=0.0
  filterred_image = converted_img.filter(ImageFilter.Kernel(size=(3, 3), kernel=kernel3x3))
  res_image = filterred_image.filter(ImageFilter.DETAIL())
  return res_image

cat = myFilter(img)
print('Изображение с фильтром:')
cat.show()
```

**2. Поясните назначение метода convert(). Поясните возможные значения параметра данного метода.** <br>
Метод convert() в библиотеке Pillow предназначен для преобразования изображения из одного цветового пространства в другое.<br>
Параметр mode метода convert() определяет цветовое пространство, в которое должно
быть преобразовано изображение. Этот параметр может принимать множество
различных значений, каждое из которых соответствует определенному цветовому
пространству. 
<br>Некоторые из наиболее распространенных значений параметра mode:<br>
'1' - черно-белый (бинарный) режим;<br>
'L' - оттенки серого (градация яркости);<br>
'RGB' - цветное изображение в формате RGB (красный, зеленый, синий);<br>
'RGBA' - цветное изображение с альфа-каналом;<br>
'CMYK' - цветное изображение в формате CMYK (циан, маджента, желтый, черный);<br>
'HSV' - цветное изображение в формате HSV (оттенок, насыщенность, значение яркости).<br>
Преобразование цветового пространства может быть полезно для многих задач
обработки изображений, таких как коррекция цвета, улучшение контраста, изменение
яркости, устранение шума и т.д.<br> <br>

**3. Что такое фильтрация изображений? Приведите примеры практического применения фильтров.**<br>

Фильтрация изображений - это процесс применения математических операций к пикселям изображения, с целью улучшения качества изображения, выделения определенных характеристик изображения или
изменения его визуального восприятия.<br>
Примеры практического применения фильтров включают в себя:
Удаление шума: фильтры могут использоваться для устранения шума с изображения, что
может быть полезно при обработке медицинских изображений или фотографий с низким
качеством.<br>
Изменение контраста: некоторые фильтры могут улучшить контрастность изображения,
что помогает при анализе изображений с малой разницей яркости.<br>
Сглаживание изображения: фильтры, такие как фильтр Гаусса, могут использоваться для
сглаживания изображения, что полезно для удаления мелких деталей.<br><br>

**4. Поясните формат изображения RGB.**<br>
RGB (Red-Green-Blue) - это цветовая модель, в которой каждый цвет на изображении
представлен комбинацией трех основных цветов: красного (Red), зеленого (Green) и
синего (Blue). В этой модели каждый пиксель изображения определяется значением
интенсивности каждого цвета, принимающим значения от 0 до 255. Например, чистый красный цвет на изображении будет представлен пикселем с RGB-значением (255, 0, 0),
чистый зеленый - (0, 255, 0), а чистый синий - (0, 0, 255).<br><br>

**5. Поясните значение терминов: растр, формат изображения, палитра, глубина, разрешение изображения.**<br>

Растр - это сетка пикселей (точек), которые составляют
изображение на экране компьютера или печатной поверхности. <br>Растровое
изображение представляет собой совокупность ячеек, каждая из которых имеет
свой цвет и яркость, определяющие цвет и яркость соответствующего пикселя на
изображении.<br>
Формат изображения - это стандарт, определяющий способ кодирования и хранения
изображения. Форматы изображений могут различаться по своим характеристикам,
таким как глубина цвета, количество кадров в секунду, сжатие и др.<br>
Палитра - это набор цветов, которые могут быть использованы для отображения
изображения. Палитра может быть ограниченной, что означает, что она содержит только
определенный набор цветов, или неограниченной, что означает, что все возможные
цвета могут быть использованы.<br>
Глубина - это количество бит, используемых для хранения цвета каждого пикселя на
изображении. Чем больше глубина цвета, тем больше цветов может быть представлено
на изображении. Например, изображения с глубиной 8 бит могут содержать 256 цветов, в
то время как изображения с глубиной 24 бита могут содержать более 16 миллионов
цветов.<br>
Разрешение изображения – величина, определяющая количество точек (элементов растрового изображения) на единицу площади (или единицу длины).
