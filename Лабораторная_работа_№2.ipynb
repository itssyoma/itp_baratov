# 2. Выполнение индивидуального задания

**Вариант 2.**
<br>
Написать скрипт для подсчета количества элементов не
целочисленного типа в списке, состоящем из элементов
произвольного типа.

Создадим список и заполним его произвольными данными разных типов

```python
p = [1846, 230, 87.189, "Dua Lipa", 6900, "я люблю тейлор свифт", (87, 23), \
     [230, 23, 46]]
```

Напишем код для перебора значений списка, чтобы узнать количество значений нецелочисленного типа.

```python
# Переменная для подсчета элементов
cnt = 0

for i in range(len(p)):
  e = p[i]
  if type(e) != int:
    print(f"{i}-й элемент: {e} | тип {type(e)} | ✓")
    cnt += 1
  else:
    print(f"{i}-й элемент: {e} | тип {type(e)}")

print("\nКоличество элементов нецелочисленного типа:", cnt)
```

**4. Что такое список в языке Python?**<br>
Списки в Python – упорядоченный изменяемый набор объектов произвольных типов, пронумерованных от 0. Они используются для хранения и работы с данными.

**5. Объясните механизм обращения к элементам списка.**<br>
Для обращения к элементам списка надо использовать индексы, которые представляют номер элемента в списка. Индексы начинаются с нуля. То есть первый элемент будет иметь индекс 0, второй элемент - индекс 1 и так далее. Для обращения к элементам с конца можно использовать отрицательные индексы, начиная с -1. То есть у последнего элемента будет индекс -1, у предпоследнего - -2 и так далее.

**6. Поясните чем список отличается от массива (например в С++).**<br>
Отличие списков от массивов заключается в том, что список - это динамическая структура, размер которого можно изменять во время выполнения программы (удалять, добавлять элементы) не задумываясь над операциями по управлению памятью (это делает транслятор). В список заносятся данные разных типов, а массив может содержать только однотипные данные.
